## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5 The total revenue is the sum total of company charges on each product sold, which is calculated in the function. The function is applied to the discount_price column to get the revenue per product and then added together.

```python
def ECom_payments(temp_dp):
    if temp_dp > 600:
        ECom_charge = temp_dp *0.25
    elif temp_dp > 350 and temp_dp <= 600 :
        ECom_charge = temp_dp *0.15
    elif temp_dp > 100 and temp_dp <= 350 :
        ECom_charge = temp_dp *0.10
    else:
        ECom_charge = temp_dp *0.05
    return ECom_charge

ECom_Data['company_revenue'] = ECom_Data['discounted_price'].apply(ECom_payments)
print('The total revenue of the E-Commerce company is :',ECom_Data['company_revenue'].sum())
```

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and 
## iv.	total BrandRevenue generated per Brand. 

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

```python
Brand_Specific = ECom_Data.copy()
Brand_Specific_1=Brand_Specific.groupby(['Brand'])['Customer_uniq_id'].count().reset_index()
Brand_Specific_2=Brand_Specific.groupby(['Brand'])['retail_price'].sum().reset_index()
Brand_Specific_3=Brand_Specific.groupby(['Brand'])['discounted_price'].sum().reset_index()
Brand_Specific_4=Brand_Specific.groupby(['Brand'])['brand_revenue'].sum().reset_index()
Brand_Specific_12 =pd.merge(Brand_Specific_1,Brand_Specific_2,how='inner',on='Brand')
Brand_Specific_34 =pd.merge(Brand_Specific_3,Brand_Specific_4,how='inner',on='Brand')
Brand_Specific_Info =pd.merge(Brand_Specific_12,Brand_Specific_34,how='inner',on='Brand')
Brand_Specific_Info.rename(columns = {'Customer_uniq_id':'count'}, inplace = True)


#Creating pair plot for the above information
sns.pairplot(Brand_Specific_Info)
plt.title("Pair plot for Brand Specifications")
plt.show()
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a) the below code gives a good visualization of the products sold in a month per region

```python
ECom_Data['Order_Date'] = pd.to_datetime(ECom_Data['Order_Date'],dayfirst=True)
Monthly_Revenue = ECom_Data.copy()
Monthly_Revenue['Order_Month']=Monthly_Revenue['Order_Date'].dt.month
Monthly_Revenue = Monthly_Revenue.groupby(['Region','Order_Month'])['company_revenue'].sum().reset_index()
plt.figure(figsize=(10,7))
sns.lineplot(data=Monthly_Revenue,x='Order_Month',y='company_revenue',hue='Region',markers=True);
plt.tight_layout()
plt.legend(bbox_to_anchor=[1,1]);
plt.title("Monthly Revenue of E-Commerce Company per region")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

### Ans 9 b) The best performing month for all regions is December(12th month), while the worst performing month varies.That is, for east the lowest can be seen in the 10th month, for north in the 7th month and stays low till 8th month ,for south it is inbetween 4th to 6th month(we can assume the 5th month ) and for west it is the 10th month.
